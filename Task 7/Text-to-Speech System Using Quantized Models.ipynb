##<B>Text-to-Speech System Using Quantized Models<B>##

```python
!apt-get update && apt-get install -y espeak espeak-ng
!pip install pyttsx3

import pyttsx3
import IPython.display as ipd

# Initialize the TTS engine
engine = pyttsx3.init()

# Function to convert text to speech and save it to a file
def text_to_speech(text, filename="output.wav"):
    engine.save_to_file(text, filename)
    engine.runAndWait()

# Sample text
text = "Hello Rahul how are you"
output_file = "output.wav"
text_to_speech(text, output_file)

# Play the generated speech
ipd.Audio(output_file)
```
