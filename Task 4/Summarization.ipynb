1. Import Necessary Libraries
The script imports the required components:
T5ForConditionalGeneration and T5Tokenizer from transformers for the T5 model.
These components handle text tokenization and summarization using the T5 model.

2. Initialize the Model and Tokenizer
Model Name: "t5-large" is specified.
Model & Tokenizer Initialization:
The model (T5ForConditionalGeneration) is loaded with pretrained weights for t5-large.
The tokenizer (T5Tokenizer) is initialized to preprocess the text as per the T5 model's requirements.

3. Define the Summarization Function
The summarize_text function performs the summarization:

  i)Input Text Preparation:
The text is prefixed with "summarize:" as T5 requires task-specific prefixes (e.g., "translate:" for translation tasks).

  ii)Tokenization:
The tokenizer converts the input text into token IDs suitable for the model.
Parameters like max_length=512 and truncation=True ensure that excessively long texts are truncated to fit the model's constraints.

 iii) Generate Summary:
The model generates summarized text using:
max_length: Limits the summary length.
min_length: Ensures the summary meets a minimum length.
length_penalty: Penalizes excessively long outputs.
num_beams: Uses beam search for generating higher-quality summaries.
early_stopping: Stops generation once an optimal output is found.

  iv)Decode Output:
The generated token IDs are converted back to human-readable text using the tokenizer.
4. Interactive Prompt for User Input
Interactive Session:
The script provides a continuous loop to accept user text until the user types 'exit'.
Summarization Execution:
The entered text is passed to the summarize_text function, and the generated summary is displayed.
Exit Condition:
If the user types 'exit', the loop terminates, and the script ends gracefully.
5. Execution Flow
Start: The script displays an initial greeting and instructions.
Input: User provides a paragraph or block of text.
Processing:
The text is processed through the tokenizer and summarization model.
Output: The generated summary is printed.
Repeat: The process repeats until the user exits.

```python
from transformers import T5ForConditionalGeneration, T5Tokenizer

# Initialize the model and tokenizer with t5-large
model_name = "t5-large"
model = T5ForConditionalGeneration.from_pretrained(model_name)
tokenizer = T5Tokenizer.from_pretrained(model_name)

def summarize_text(text, max_length=150):
    # Preprocess the text for T5 (add "summarize:" prefix)
    input_text = "summarize: " + text
    inputs = tokenizer(input_text, return_tensors="pt", max_length=512, truncation=True)

    # Generate summary
    summary_ids = model.generate(inputs["input_ids"], max_length=max_length, min_length=30, length_penalty=2.0, num_beams=4, early_stopping=True)
    summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
    return summary

# Interactive prompt to get user input
if __name__ == "__main__":
    print("T5 Summarization System")
    print("Enter the text you want to summarize (or type 'exit' to quit):")

    while True:
        # Get user input
        user_input = input("Your Text: ")

        # Check for exit condition
        if user_input.lower() == "exit":
            print("Exiting the summarization system. Goodbye!")
            break

        # Summarize and print the result
        summary = summarize_text(user_input)
        print("\nSummary:", summary)
        print("\n" + "-"*50 + "\n")
```
